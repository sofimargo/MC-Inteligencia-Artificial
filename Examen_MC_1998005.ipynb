```python
#Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cargar datos desde un archivo csv
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
```

```python
#Muestra las primeras 5 filas de datos
print(df.head(5))
```

```python
#Información de los tipos de datos de cada columna
print(df.info())
```

```python
#Verificar si hay valores nulos en la columna horsepower
print(df.isnull().sum())
```

```python
#Reemplaza esos valores nulos con la mediana de esa misma columna
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].median())
```

```python
#Crear nueva columna
df['relacion_peso_potencia']=df['weight']/df['horsepower']
```

```python
#Verificar si hay valores nulos
print(df.isnull().sum())
```

```python
#Histograma: Muestra la distribución de la aceleración (acceleration) de los vehículos
plt.hist(df['acceleration'], bins=20, color= 'red', edgecolor='black')
plt.title('Distribución de la aceleración')
plt.xlabel('Aceleración(m/s^2)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Gráfico de Dispersión (Scatter): Grafica el Desplazamiento del motor (displacement) en el
#eje X versus los Caballos de fuerza (horsepower) en el eje Y.
plt.scatter(df["displacement"], df["horsepower"], alpha=0.5, color='purple')
plt.title("Gráfica de dispersión de desplazamiento vs caballos de fuerza")
plt.xlabel("Desplazamiento")
plt.ylabel("Caballos de fuerza")
plt.show()
```

```python
#Gráfico de Barras: Muestra el promedio de rendimiento (mpg - millas por galón)
#agrupado por el país de origen (origin).
mpg_agrupado = df.groupby("origin")["mpg"].mean()
mpg_agrupado.plot(kind="bar", color=['red','pink','blue'])
plt.title("Gráfico de barras rendimiento promedio")
plt.xlabel("OPaís de origen")
plt.ylabel("Promedio MPG")
plt.xticks(rotation=0)
plt.show()
```
